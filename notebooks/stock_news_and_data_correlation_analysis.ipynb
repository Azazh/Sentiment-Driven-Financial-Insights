```python
import sys
import os

# Add the parent directory to the Python path
sys.path.append(os.path.abspath(".."))
# Import necessary libraries
from scripts.stock_news_and_data_correlation_analysis import StockSentimentAnalyzer

# Define file paths
stock_data_directory = "../data/yfinance_data/"
news_data_file = "../data/raw_analyst_ratings/raw_analyst_ratings.csv"

# Initialize the analyzer
analyzer = StockSentimentAnalyzer(stock_data_path=stock_data_directory, news_data_path=news_data_file)

# Analyze all stocks
results = analyzer.analyze_all_stocks()

# Display correlation results
print("Sentiment and Stock Movement Correlation Results:")
for stock, result in results.items():
    print(f"{stock}: Correlation={result['Correlation']:.4f}, P-Value={result['P-Value']:.4f}")
```
